# Clusterização usando K-Means
#### Aplicando o algoritmo de clusterização com valores gerados aleatoriamente

```python
# importando as bibliotecas no python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.datasets.samples_generator import make_blobs
```

```python
# criando os dados gerados aleatoriamente
dados, clusters = make_blobs(n_samples = 1000, centers = 4)

# configurando o gráfico
plt.figure(figsize = (10, 5))
plt.title('Dados Gerados Aleatoriamente')
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(dados[:, 0], dados[:, 1], marker = 'x', color = 'green', label = 'Dados')
plt.grid(True)
plt.box(False)
plt.legend()
```

```python
# criando o clusterizados 'kmeans'
kmeans = KMeans(n_clusters = 4)
```

```python
# treinando o clusterizados 'kmeans'
kmeans.fit(dados)
```

```python
# realizando as previsoes com 4 clusterizadores para os dados gerados
previsoes = kmeans.predict(dados)
```

```python
# configurando o gráfico
plt.figure(figsize = (10, 5))
plt.title('Dados Clusterizados')
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(dados[:, 0], dados[:, 1], marker = 'x', c = previsoes, label = 'Dados')
plt.grid(True)
plt.box(False)
plt.legend()
```
