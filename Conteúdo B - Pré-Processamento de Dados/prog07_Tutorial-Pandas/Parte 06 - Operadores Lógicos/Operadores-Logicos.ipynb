# Usando Operadores Lógicos

Nesse arquivo, será usado operadores lógicos para fazer pré-processamento de dados usando a linguagem de programação Python.

```python
import pandas as pd
# importando a biblioteca "Pandas" do Python
```

```python
dados = pd.read_csv("bank.csv", encoding = "UTF-8", sep = ";")
# atribuindo ao objeto "dados" todos os registros do banco de dados presentes no arquivo "bank.csv"
```

```python
dados
# visualizando os registros armazenados no objeto "dados"
```

Podemos fazer uma seleção por campos de dados através dos registros. No exemplo abaixo, será selecionado somente as pessoas que possuem "single" armazenadas no registro "marital".

```python
dados[dados.marital == "single"]
# fazendo operações lógicas pelo Pandas. No exemplo abaixo, será selecionado somente as pessoas que possuem
# "single" armazenadas no registro "marital".
```

Agora vamos propor a seguinte situação: queremos ver a idade e o emprego das pessoas que são solteiras e possuem o primário concluído. Observe abaixo o uso do operador lógico & ou "e" para fazer essa operação.

```python
dados[(dados.marital == "single") & (dados.education == "primary")][["age","job"]]
# fazendo a visualização da idade e o emprego das pessoas que são solteiras e possuem o primário concluído
```

Desse modo, podemos fazer seleções de dados de forma mais precisa, usando o atributo "filter".
