### Usando Diversos Algoritmos de Agrupamento para Comparar os Seus Resultados

```python
# importando a biblioteca matplotlib do python
import matplotlib.pyplot as plt
%matplotlib inline

# importando a biblioteca numpy do python
import numpy as np

# importando a biblioteca sklearn do python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn import datasets
```

```python
# importando um dataset do sklearn com os seus respectivos clusters
dataframe, cluster_original = datasets.make_moons(n_samples = 1500, noise = 0.09)
# 'make_moons' dados em formato de luas
# 'noise' adicionar ru√≠do aos dados
```

```python
# plotando os dados dispon√≠veis
plt.figure(figsize = (12, 7))
plt.title('Dados n√£o Clusterizados')
plt.scatter(dataframe[:, 0], dataframe[: , 1], marker = 'x', label = 'Dados')
plt.grid(True)
plt.box(False)
plt.legend()
```

```python
# criando uma lista de cores para os clusters
cores = np.array(['red', 'blue'])
```

### Algoritmo k-means

```python
# definindo o clusterizados 'kmeans'
kmeans = KMeans(n_clusters = 2)
```

```python
# realizando o treinamento e as previs√µes com o clusterizados
previsoes = kmeans.fit_predict(dataframe)
```

```python
# plotando os dados dispon√≠veis
plt.figure(figsize = (12, 7))
plt.title('Dados Clusterizados (K-Means)')
plt.scatter(dataframe[:, 0], dataframe[: , 1], marker = '*', color = cores[previsoes])
plt.grid(True)
plt.box(False)
```

### Agrupamento Hier√°rquico

```python
# definindo o clusterizados 'hc'
hc = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
```

```python
# realizando o treinamento e as previs√µes com o clusterizados
previsoes = hc.fit_predict(dataframe)
```

```python
# plotando os dados dispon√≠veis
plt.figure(figsize = (12, 7))
plt.title('Dados Clusterizados (Agrupamento Hier√°rquico)')
plt.scatter(dataframe[:, 0], dataframe[: , 1], marker = '*', color = cores[previsoes])
plt.grid(True)
plt.box(False)
```

### Algoritmo DBSCAN

```python
# definindo o clusterizados 'dbscan'
dbscan = DBSCAN(eps = 0.1)
```

```python
# realizando o treinamento e as previs√µes com o clusterizados
previsoes = dbscan.fit_predict(dataframe)
```

```python
# plotando os dados dispon√≠veis
plt.figure(figsize = (12, 7))
plt.title('Dados Clusterizados (Agrupamento Hier√°rquico)')
plt.scatter(dataframe[:, 0], dataframe[: , 1], marker = '*', color = cores[previsoes])
plt.grid(True)
plt.box(False)
```
