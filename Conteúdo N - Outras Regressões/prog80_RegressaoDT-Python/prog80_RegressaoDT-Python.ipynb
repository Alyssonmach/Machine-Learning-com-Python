# Regressão - Árvores de Decisão
#### Usando algoritmo de árvores de decisão para implentar em um problema de regressão e não linear para preço de planos de saúde.

```python
# importando a biblioteca pandas do python
import pandas as pd
```

```python
# implementando a base de dados no dataframe
dataframe = pd.read_csv('plano_saude2.csv', encoding = 'utf-8', sep = ',')
```

```python
# separando as variáveis x e y
x = dataframe.iloc[:, 0:1].values
y = dataframe.iloc[:, 1].values
```

```python
# importando a biblioteca sklearn do python
from sklearn.tree import DecisionTreeRegressor
```

```python
# criando o classificador 'regressor'
regressor = DecisionTreeRegressor()
```

```python
# treinando o classificador
regressor.fit(x, y)
```

```python
# visualizando o quão o modelo feito se adaptou aos dados
score = regressor.score(x, y)
print(score)
```

Observe que o modelo criado se adaptou 100% aos dados disponíveis.

```python
# importando a biblioteca matplotlib do python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# configurando o gráfico
plt.figure(figsize = (10,5))
plt.scatter(x, y, color = 'red', label = 'Dados')
plt.plot(x, regressor.predict(x), color = 'green', label = 'Regressão com Árvores de Decisão')
plt.title('Regressão com Árvores de Decisão')
plt.xlabel('Idade')
plt.ylabel('Custo')
plt.legend()
plt.grid(True)
plt.box(False)
```

```python
# importando a biblioteca numpy do python
import numpy as np
```

```python
# criando uma lista que começa com a menor variável x, até a maior, variando na escala 0.1
x_teste = np.arange(min(x), max(x), 0.1)
# alinhando em um vetor coluna
x_teste = x_teste.reshape(-1, 1)
```

```python
# configurando o gráfico
plt.figure(figsize = (10,5))
plt.scatter(x, y, color = 'red', label = 'Dados')
plt.plot(x_teste, regressor.predict(x_teste), color = 'green', label = 'Regressão com Árvores de Decisão')
plt.title('Regressão com Árvores de Decisão')
plt.xlabel('Idade')
plt.ylabel('Custo')
plt.legend()
plt.grid(True)
plt.box(False)
```

```python
# fazendo a predição para uma pessoa com 40 anos de idade
regressor.predict([[40]])
```

Observe que o preço do plano de saúde foi de 1150 reais para uma pessoa de 40 anos de idade.
