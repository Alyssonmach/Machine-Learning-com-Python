# Regressão - Vetores de Suporte
#### Usando o algoritmo de vetores de suporte em regressão para um problema não linear.

```python
# importando a biblioteca pandas do python
import pandas as pd
```

```python
# colocando os dados no dataframe
dataframe = pd.read_csv('plano_saude2.csv', encoding = 'utf-8', sep = ',')
```

```python
# separando as variáveis x e y
x = dataframe.iloc[:, 0:1].values
y = dataframe.iloc[:, 1:2].values
```

### Kernel Linear

```python
# importando a biblioteca sklearn do python
from sklearn.svm import SVR
```

```python
# criando o classificador 'regressor_linear'
regressor_linear = SVR(kernel = 'linear')
```

```python
# treinando o classificador
regressor_linear.fit(x, y)
```

```python
# importando a biblioteca matplotlib do python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# configurando o gráfico
plt.figure(figsize = (10,5))
plt.scatter(x, y, color = 'red', label = 'Dados')
plt.plot(x, regressor_linear.predict(x), color = 'green', label = 'Modelo Vetores de Suporte')
plt.title('Modelo de Regressão Vetores de Suporte')
plt.xlabel('Idade')
plt.ylabel('Custo')
plt.legend()
plt.grid(True)
plt.box(False)
```

```python
# visualizando a porcentagem de adaptação do modelo feito em relação aos dados
score1 = regressor_linear.score(x, y)
print(score1)
```

### Kernel Poly

```python
# criando o classificador 'regressor_poly'
regressor_poly = SVR(kernel = 'poly', degree = 3)
```

```python
# treinando o classificador
regressor_poly.fit(x, y)
```

```python
# configurando o gráfico
plt.figure(figsize = (10,5))
plt.scatter(x, y, color = 'red', label = 'Dados')
plt.plot(x, regressor_poly.predict(x), color = 'green', label = 'Modelo Vetores de Suporte')
plt.title('Modelo de Regressão Vetores de Suporte')
plt.xlabel('Idade')
plt.ylabel('Custo')
plt.legend()
plt.grid(True)
plt.box(False)
```

```python
# visualizando a porcentagem de adaptação do modelo feito em relação aos dados
score2 = regressor_poly.score(x, y)
print(score2)
```

### Kernel rbf

```python
# criando o classificador 'regressor_rbf'
regressor_rbf = SVR(kernel = 'rbf')
```

```python
# importando a biblioteca sklearn do python
from sklearn.preprocessing import StandardScaler
```

```python
# criando o objeto 'scaler_x'
scaler_x = StandardScaler()
```

```python
# realizando o escoalonamento nas variáveis x
x = scaler_x.fit_transform(x)
```

```python
# criando o objeto 'scaler_y'
scaler_y = StandardScaler()
```

```python
# realizando o escoalonamento nas variáveis y
y = scaler_y.fit_transform(y)
```

```python
# criando o classificador 'regressor_rbf'
regressor_rbf = SVR(kernel = 'rbf')
```

```python
# treinando o classificador
regressor_rbf.fit(x, y)
```

```python
# configurando o gráfico
plt.figure(figsize = (10,5))
plt.scatter(x, y, color = 'red', label = 'Dados')
plt.plot(x, regressor_rbf.predict(x), color = 'green', label = 'Modelo Vetores de Suporte')
plt.title('Modelo de Regressão Vetores de Suporte')
plt.xlabel('Idade')
plt.ylabel('Custo')
plt.legend()
plt.grid(True)
plt.box(False)
```

```python
# visualizando a porcentagem de adaptação do modelo feito em relação aos dados
score3 = regressor_rbf.score(x, y)
print(score3)
```
